```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("raw/loan/loan.csv")
df
```

#### Обработка данных
Среди признаков можно выделить несколько категориальных, для дальнейшей обработки стоит заменить придать им числовые значения: 
1. EmploymentStatus: [Employed, Self-Employed, Unemployed, ...] 
2. EducationLevel: [Master, Associate, Bachelor, High School, ...]
3. MaritalStatus: [Single, Married, Divorced, ...]
4. HomeOwnershipStatus: [Own, Mortgage, Rent, Other, ...]
5. LoanPurpose: [Home, Debt Consolidation, Other, ...]

Так же можно удалить столбец даты ApplicationDate, так как в этой выборке данных он не влияет на итоговый рейтинг


```python
# Drop dates and encode labels
df = df.drop(columns=["ApplicationDate"]) if "ApplicationDate" in df.columns else df
categorical_cols = [
    "EmploymentStatus", "EducationLevel", "MaritalStatus", "HomeOwnershipStatus", "LoanPurpose"
]
for col in categorical_cols:
    if isinstance(df.dtypes[col], np.dtypes.ObjectDType):  # Do mapping if col contains strings
        mapping = {label: i for i, label in enumerate(df[col].unique())}
        df[col] = df[col].map(mapping)
        print(f"Отображение {col}: {mapping}")

df
```

```python
# Print averages and standard deviation
df.describe(include="all")
```

```python
# Research the numerical features of a data set
numeric_df = df.select_dtypes(include=[np.number])

# Build histograms plot
def plot_hist(data, **kwargs):
    # Bins and discrete are not used, as they critically slow down processing
    sns.histplot(data=data, kde=True, **kwargs)

g = sns.FacetGrid(pd.melt(numeric_df), col="variable", col_wrap=3, sharex=False, sharey=False)
g.map_dataframe(plot_hist)
g.set_titles("{col_name} histogram")
g.set_axis_labels("Value", "Frequency")
plt.show()
```

Из графиков и таблиц видно, что выбросов и пропусков нет, что логично, ведь набор данных создан программно

Изучим корреляцию данных

```python
# %matplotlib notebook

# Abs threshold value for display on heatmap (0.0 up to 1.0) 
# Set THRESHOLD = 0 if no filtration needed
THRESHOLD = 0.3

corr_matrix = numeric_df.corr()
corr_matrix = corr_matrix.mask(abs(corr_matrix) < THRESHOLD, np.nan)

plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
corr_matrix
```

Из корреляционной матрицы можно увидеть средние (0.3 - 0.5) и сильные (0.5 - 1.0) зависимости следующих признаков:

```python
# Age to Experience correlation research
plt.figure(figsize=(10, 8))
sns.scatterplot(x="Age", y="Experience", data=df, color="gray", alpha=0.3, s=30)
sns.lineplot(x="Age", y="Experience", data=df, color="magenta")

plt.xlabel("Age")
plt.ylabel("Experience")
plt.title("Age to Experience")
plt.show()
```

```python
plt.figure(figsize=(20, 12))

# First plot: Age - CreditScore
plt.subplot(2, 2, 1)
sns.scatterplot(x="Age", y="CreditScore", data=df, color="blue", label="Age vals")
sns.regplot(x="Age", y="CreditScore", data=df, scatter=False, color="red", label="Age linear")
plt.title("CreditScore to Age", fontsize=14)
plt.legend()

# Second plot: Experience - CreditScore
plt.subplot(2, 2, 2)
sns.scatterplot(x="Experience", y="CreditScore", data=df, color="green", label="Exp vals")
sns.regplot(x="Experience", y="CreditScore", data=df, scatter=False, color="magenta", label="Exp linear")
plt.title("CreditScore to Experience", fontsize=14)
plt.legend()

# Interpolate Experience to Age for a visual comparison
exp_interp = np.interp(
    df['Experience'], 
    (df['Experience'].min(), df['Experience'].max()), 
    (df['Age'].min(), df['Age'].max())
)

# Third plot: Experience + Age to CreditScore
plt.subplot(2, 1, 2)
sns.scatterplot(x="Age", y="CreditScore", data=df, color="blue", label="Age vals")
sns.regplot(x="Age", y="CreditScore", data=df, scatter=False, color="red", label="Age linear")
sns.scatterplot(x=exp_interp, y="CreditScore", data=df, color="green", label="Experience vals")
sns.regplot(x=exp_interp, y="CreditScore", data=df, scatter=False, color="yellow", label="Experience linear")

plt.xlabel("Age & Experience", fontsize=12)
plt.ylabel("CreditScore", fontsize=12)
plt.title("Overlaying Age and Experience graphs on CreditScore", fontsize=14)
plt.legend()

# Show
plt.tight_layout()
plt.show()
```

```python
# Monthly to Annual AND Rate to Base Rate

# First plot: MonthlyIncome to AnnualIncome
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x="MonthlyIncome", y="AnnualIncome", data=df, color="blue")
plt.title("MonthlyIncome to AnnualIncome")

# Second plot: InterestRate to BaseInterestRate
plt.subplot(1, 2, 2)
sns.scatterplot(x="BaseInterestRate", y="InterestRate", data=df, color="blue")
sns.regplot(x="BaseInterestRate", y="InterestRate", data=df, scatter=False, color="red")
plt.title("InterestRate to BaseInterestRate")

# Show
plt.tight_layout()
plt.show()
```

**Выводы**
1. Кредитный рейтинг растёт пропорционально возрасту (опыту) и с ростом сужается диапазон принимаемых значений
2. Age и Experience имеют отклонения только в граничных значениях. Среди молодых кто-то раньше начинает работать, среди взрослых кто-то раньше заканчивает. Можно опираться на Experience
3. AnnualIncome действительно зависит от MonthlyIncome и имеет множество выбросов, но для кредита важен краткосрочный стабильный доход, поэтому можно оставить MonthlyIncome 
4. InterestRate зависит от BaseInterestRate и находится в определенном диапазоне значений. Фактическая ставка важнее для кредита, поэтому базовую можно удалить
